```python
import ROOT
%jsroot on
from ROOT import RooFit, RooWorkspace
```

```python
mass = ROOT.RooRealVar("mass", "mass", 5.20, 5.30)
mean = ROOT.RooRealVar("mean", "mean", 5.28, 5.20, 5.30)
sigma = ROOT.RooRealVar("sigma", "sigma", 0.005, 0.001, 0.05)
```

```python
model = ROOT.RooGaussian("model", "model", mass, mean, sigma)
```

```python
# Criar o workspace e importar o modelo
w = ROOT.RooWorkspace("w")
getattr(w, "import")(model)
# 'getattr' Ã© usado para contornar o uso de 'import', palavra-chave em Python
```

```python
# Gerar 1000 eventos a partir da PDF
dado = w.pdf("model").generate(ROOT.RooArgSet(w.var("mass")), 1000)
w.pdf("model").fitTo(dado)
```

```python
frame = w.var("mass").frame()
dado.plotOn(frame)
w.pdf("model").plotOn(frame)

w.writeToFile("wspace.root")

c = ROOT.TCanvas("c", "c", 900, 700)
frame.Draw()
c.Draw()
c.SaveAs("w.png")
```
